# Approach

1. **Load the Dataset**  
   Load the dataset from the website that provides an online catalog of all known confirmed exoplanets detected using the **Radial Velocity** method.

2. **Filter the Dataset**  
   Filter the dataset to include only those exoplanets with orbital periods between **0 and 400 Earth days** in order to plot the histogram.

3. **Generate the Histogram**  
   Use the Python library **matplotlib** to generate the histogram.

4. **Analyze the Distribution**  
   Analyze the histogram to detect various insights about the orbital periods of the exoplanets.


```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = './dataset.csv'
data = pd.read_csv(file_path)
data
```

```python
# Filtering data from the dataset
filtered_data = data[(data['orbital_period'] > 0) & (data['orbital_period'] <= 400)]
filtered_data
```

```python
#Ploting histogram
plt.figure(figsize=(10, 6))
plt.hist(filtered_data['orbital_period'], bins=40, color='skyblue', edgecolor='black')
plt.title('Distribution of Exoplanets\' Orbital Periods')
plt.xlabel('Orbital Period (Earth days)')
plt.ylabel('Number of Exoplanets')
plt.grid(True)
plt.savefig('exoplanet_orbital_periods_histogram.png', format='png')
plt.show()
```

```python
# Percentage of exoplanets with orbital period less than 50 days
less_than_50_days = filtered_data[filtered_data['orbital_period'] < 50]
percentage_less_than_50 = (len(less_than_50_days) / len(filtered_data)) * 100

print(f"Percentage of exoplanets with orbital period less than 50 Earth days: {percentage_less_than_50:.2f}%")
```

### The histogram shows a significant concentration of exoplanets with orbital periods less than 50 Earth days 59.67% in particular.

```python
# Basic statistical analysis
mean_period = filtered_data['orbital_period'].mean()
median_period = filtered_data['orbital_period'].median()
mode_period = filtered_data['orbital_period'].mode()[0]
std_dev_period = filtered_data['orbital_period'].std()
percentiles = filtered_data['orbital_period'].quantile([0.25, 0.5, 0.75])

print(f"Mean: {mean_period}")
print(f"Median: {median_period}")
print(f"Mode: {mode_period}")
print(f"Standard Deviation: {std_dev_period}")
print("25th, 50th, and 75th Percentiles:")
for percentile, value in zip(['25th', '50th', '75th'], percentiles):
    print(f"{percentile}: {value:.2f}")
```

```python
#identifying outliers
Q1 = filtered_data['orbital_period'].quantile(0.25)
Q3 = filtered_data['orbital_period'].quantile(0.75)
IQR = Q3 - Q1
outliers = filtered_data[(filtered_data['orbital_period'] < (Q1 - 1.5 * IQR)) | (filtered_data['orbital_period'] > (Q3 + 1.5 * IQR))]

print(f"Number of Outliers: {len(outliers)}")
```

# Observations

### 1. **Mean (86.03)**
The **mean** is the average value of the orbital periods in the dataset. This value is significantly higher than both the **median** (29.22) and the **mode** (2.5), which suggests that there may be a few exoplanets with very large orbital periods that are pulling the mean upward. ### 2. **Median (29.22)**

The fact that the median is much lower than the mean suggests a **right-skewed distribution**, meaning there are more exoplanets with smaller orbital periods (closer to 2.5 days) than those with very large orbital periods.

### 3. **Mode (2.5)**
The **mode** represents the most frequently occurring value in the dataset. A mode of **2.5** suggests that many of the exoplanets have an orbital period around 2.5 days. This may indicate a cluster of exoplanets with relatively short orbital periods.

### 4. **Standard Deviation (110.14)**
The **standard deviation** is a measure of the spread or dispersion of the data points from the mean. A large standard deviation (**110.14**) relative to the mean (**86.03**) indicates a high degree of variability in the orbital periods. This confirms the presence of significant outliers or exoplanets with very large orbital periods, which are causing the data to spread widely.

### 5. **Outliers (52)**
There are **52 outliers** in the dataset. These could represent rare but scientifically interesting cases, such as planets in long-period orbits or those affected by specific gravitational interactions. Having 52 outliers suggests that the dataset contains a significant number of data points that deviate substantially from the typical range of orbital periods.


# Detailed Analysis of Exoplanet Orbital Period Distribution

## 1. **Concentration of Short-Period Exoplanets (Less than 50 Earth Days)**

The histogram reveals a significant concentration of exoplanets with orbital periods of less than 50 Earth days. Several factors contribute to this:

### Key Factors:

- **Radial Velocity Method Sensitivity**:  
  Short-period exoplanets are closer to their host stars, leading to more frequent and larger Doppler shifts. This makes them easier to detect compared to long-period planets.

- **Observational Limitations**:  
  The relatively short observation window for exoplanet detection favors short-period planets, which complete their orbits quickly and can be observed multiple times.

- **Selection Bias**:  
  Detection methods, such as radial velocity, are inherently biased toward short-period planets, as they induce a more pronounced gravitational effect on their stars.

- **Nature of Planetary Orbits**:  
  Close-in orbits are more common, especially in systems with smaller, lower-mass stars, due to planetary migration during system formation.

## 2. **Scarcity of Long-Period Exoplanets (Several Hundred Days)**

There is a marked scarcity of exoplanets with orbital periods in the several hundred-day range, similar to Earth, Venus, or Mars. This scarcity is attributed to several factors:

### Key Challenges:

- **Smaller Doppler Shifts**:  
  Planets on longer orbits are farther from their stars, causing smaller gravitational tugs and weaker Doppler shifts that are harder to detect.

- **Extended Observation Periods**:  
  Long-period planets require multi-year observation campaigns, making them harder to detect within the typical observation timeframe.

- **Technological Limitations**:  
  Current detection methods struggle to identify small amplitude shifts associated with planets on long orbits, particularly those akin to Earth.

- **Habitable Zone Planets**:  
  Many long-period planets fall within the habitable zone of their stars, where conditions could be suitable for life. However, the detection of such planets remains difficult with existing methods.

## **Implications for Exoplanet Research**

### 1. **Detection Bias**:  
  The overrepresentation of short-period planets highlights the inherent bias of current detection methods. Addressing this bias is crucial for discovering more Earth-like planets.

### 2. **Technological Advancements**:  
  Future improvements in instrumentation, such as more precise spectrographs, and longer-term observational missions will be vital in detecting long-period, Earth-like exoplanets.

### 3. **Search for Habitable Worlds**:  
  The current limitations underscore the need for complementary detection methods, such as transit photometry and direct imaging, to identify Earth-like planets within habitable zones.

This analysis emphasizes the critical need for advancing detection technologies and methodologies to balance the current distribution of known exoplanets, allowing for a more comprehensive search for potentially habitable worlds.
